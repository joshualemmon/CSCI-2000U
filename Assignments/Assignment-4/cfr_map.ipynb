```python
%pylab inline
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

def mostCases(list):
    largest = list['Cases'][0]
    for i in range(len(list)):
        if list['Cases'][i] > largest:
            largest = list['Cases'][i]
    return largest
    

file = open("./Source_data_for_CFR_vaccine_map_abridged.csv", "rt")
try:
    reader = csv.reader(file)
    data = []
    for row in reader:
        data.append(row)
finally:
    file.close()
E = np.zeros(len(data), dtype=[('Category', str, 2),
                               ('Position', float, 2),
                               ('Cases', int, 1)])
measlescount = 0
mumpscount = 0
rubellacount = 0
poliocount = 0

for i in range(len(data)-1):
    E['Category'][i] = data[i+1][0]
    if data[i+1][0] == "Measles":
        measlescount +=1
    elif data[i+1][0] == "Mumps":
        mumpscount += 1
    elif data[i+1][0] == "Rubella":
        rubellacount += 1
    elif data[i+1][0] == "Polio":
        poliocount += 1
    data[i+1][2] = data[i+1][2].replace('°','')
    data[i+1][1] = data[i+1][1].replace('°','')
    E['Position'][i] = float(data[i+1][1]),float(data[i+1][2])
    #some case values are seperated by commas eg 131,441
    data[i+1][3] = data[i+1][3].replace(',', '')
    if data[i+1][3] == 'unknown' or data[i+1][3] == '' or data[i+1][3] == 'Unknown':
        E['Cases'][i] = 0;
    else:
        E['Cases'][i] = int(data[i+1][3])

measles = np.zeros(measlescount, dtype=[('Position', float, 2), ('Cases', int, 1)])
mumps = np.zeros(mumpscount, dtype=[('Position', float, 2), ('Cases', int, 1)])
rubella = np.zeros(rubellacount, dtype=[('Position', float, 2), ('Cases', int, 1)])
polio = np.zeros(poliocount, dtype=[('Position', float, 2), ('Cases', int, 1)])

measlescount = 0
mumpscount = 0
rubellacount = 0
poliocount = 0
for i in range(len(E)):
    if E[i][0] == "Me":
        measles[measlescount][0] = E[i][1]
        measles[measlescount][1] = E[i][2]
        measlescount +=1
    elif E[i][0] == "Mu":
        mumps[mumpscount][0] = E[i][1]
        mumps[mumpscount][1] = E[i][2]
        mumpscount += 1
    elif E[i][0] == "Ru":
        rubella[rubellacount][0] = E[i][1]
        rubella[rubellacount][1] = E[i][2]
        rubellacount += 1
    elif E[i][0] == "Po":
        polio[poliocount][0] = E[i][1]
        polio[poliocount][1] = E[i][2]
        poliocount += 1

mostmeasles = int(mostCases(measles))
mostmumps = int(mostCases(mumps))
mostrubella = int(mostCases(rubella))
mostpolio = int(mostCases(polio))

print(polio['Position'][0])

fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(111)
ax.set_title("Measles outbreak")

measlesmap = Basemap(projection='mill')

measlesmap.drawcoastlines(color='0.50', linewidth=0.25)
measlesmap.fillcontinents(color='0.95')

for x in range(len(measles)):
    cases = measles['Cases'][x]
    measlesmap.plot(measlesmap(*measles['Position'][x]), 'bo', markersize= 20*cases/mostmeasles)

fig2 = plt.figure(figsize=(14,10))
ax2 = fig2.add_subplot(111)
ax2.set_title("Mumps outbreak")

mumpsmap = Basemap(projection='mill')

mumpsmap.drawcoastlines(color='0.50', linewidth=0.25)
mumpsmap.fillcontinents(color='0.95')

for x in range(len(mumps)):
    cases = mumps['Cases'][x]
    mumpsmap.plot(mumpsmap(*mumps['Position'][x]), 'bo', markersize= 20*cases/mostmumps)

fig3 = plt.figure(figsize=(14,10))
ax3 = fig3.add_subplot(111)
ax3.set_title("Rubella outbreaks")
ax3.set_xticks([0,1,2,3])

rubellamap = Basemap(projection='mill')

rubellamap.drawcoastlines(color='0.50', linewidth=0.25)
rubellamap.fillcontinents(color='0.95')

for x in range(len(rubella)):
    cases = rubella['Cases'][x]
    rubellamap.plot(rubellamap(*rubella['Position'][x]), 'bo', markersize= 20*cases/mostrubella)

fig4 = plt.figure(figsize=(14,10))
ax4 = fig4.add_subplot(111)
ax4.set_title("Polio outbreaks")

poliomap = Basemap(projection='mill')

poliomap.drawcoastlines(color='0.50', linewidth=0.25)
poliomap.fillcontinents(color='0.95')

for x in range(len(polio)):
    cases = polio['Cases'][x]
    poliomap.plot(poliomap(*polio['Position'][x]), 'bo', markersize= 20*cases/mostpolio)

plt.show()
```
